```python
import cv2
import dlib
from deepface import DeepFace
import pandas as pd

# Initialize dlib's face detector
detector = dlib.get_frontal_face_detector()

# Initialize webcam
cap = cv2.VideoCapture(1)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    # Detect faces using dlib
    faces = detector(frame)

    for face in faces:
        # Get the coordinates of the face
        x1, y1, x2, y2 = face.left(), face.top(), face.right(), face.bottom()

        # Crop the face region
        face_crop = frame[y1:y2, x1:x2]

        try:
            # Perform face recognition using DeepFace
            result = DeepFace.find(face_crop, db_path="/Users/komangandikawirasantosa/Face-Rec/DeepFace-IPYNB/Database",
                                   detector_backend="dlib",model_name="Facenet512")

            # Draw a rectangle around the face
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Check if the result is not empty
            if isinstance(result, pd.DataFrame) and not result.empty:
                # Extract the person's name from the first row of the DataFrame
                person_name = result.iloc[0]["identity"].split('/')[-2]  # Extract person name from path
                cv2.putText(frame, person_name, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
            else:
                print("No match found.")

        except Exception as e:
            print(f"Error during face recognition: {str(e)}")

    # Display the resulting frame
    cv2.imshow('Face Detection and Recognition', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```python
from deepface import DeepFace
import matplotlib.pyplot as plt
import numpy as np
import cv2

# List of available backends, models, and distance metrics
backends = ["opencv", "ssd", "dlib", "mtcnn", "retinaface"]
models = ["VGG-Face", "Facenet", "Facenet512", "OpenFace", "DeepFace", "DeepID", "ArcFace", "Dlib", "SFace"]
metrics = ["cosine", "euclidean", "euclidean_l2"]

img = "images/patricia.jpg"

def face_recognition(img):
    # Perform face recognition on the provided image
    people = DeepFace.find(img_path=img, db_path="images/", model_name=models[0], distance_metric=metrics[1])

    # Display the original image
    plt.imshow(cv2.imread(img))
    plt.axis('off')  # Hide axes
    plt.show()

    # Print the identities of the recognized people
    for person in people:
        identity = person.get('identity', ['Unknown'])[0]
        name = identity.split('/')[-1].split('.')[0]  # Extract name without extension
        print(name)

def realtime_face_recognition():
    # Define a video capture object
    vid = cv2.VideoCapture(1)

    while True:
        # Capture the video frame by frame
        ret, frame = vid.read()
        if not ret:
            print("Failed to capture video frame.")
            break

        # Perform face recognition on the captured frame
        try:
            people = DeepFace.find(img_path=frame, db_path="/Users/komangandikawirasantosa/Face-Rec/DeepFace-IPYNB/Database", model_name=models[0], distance_metric=metrics[2], enforce_detection=False)
        except Exception as e:
            print(f"Error during face recognition: {e}")
            continue

        # Check if the people list is empty
        if not people:
            cv2.putText(frame, "Unknown", (50, 50), cv2.FONT_ITALIC, 2, (0, 0, 255), 2)
        else:
            for person in people:
                # Ensure required keys are in the person dictionary
                if all(key in person for key in ['source_x', 'source_y', 'source_w', 'source_h']):
                    # Retrieve the coordinates of the face bounding box safely
                    x = person['source_x'][0] if len(person['source_x']) > 0 else 0
                    y = person['source_y'][0] if len(person['source_y']) > 0 else 0
                    w = person['source_w'][0] if len(person['source_w']) > 0 else 0
                    h = person['source_h'][0] if len(person['source_h']) > 0 else 0

                    # Draw a rectangle around the face
                    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

                    # Get the person's name and display it on the image
                    if 'identity' in person and len(person['identity']) > 0:
                        identity = person['identity'][0].split('/')[-1].split('.')[0]  # Extract name without extension
                        cv2.putText(frame, identity, (x, y), cv2.FONT_ITALIC, 1, (0, 0, 255), 2)
                    else:
                        cv2.putText(frame, "Unknown", (x, y), cv2.FONT_ITALIC, 1, (0, 0, 255), 2)

        # Display the resulting frame
        cv2.namedWindow('frame', cv2.WINDOW_NORMAL)
        cv2.resizeWindow('frame', 960, 720)
        cv2.imshow('frame', frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    vid.release()
    cv2.destroyAllWindows()

# Perform face recognition on a single image
# face_recognition(img)

# Perform real-time face recognition using the webcam
realtime_face_recognition()
```

```python
import time
from deepface import DeepFace
import cv2

# Optimized variables
backends = ["mtcnn", "retinaface"]  # Faster backends
models = ["Facenet", "ArcFace"]     # Efficient models for faster recognition
metrics = ["cosine", "euclidean"]   # Metrics for distance calculation
frame_count = 0
process_every_n_frame = 5  # Process every 5th frame for speed

def realtime_face_recognition():
    # Define a video capture object
    vid = cv2.VideoCapture(0)

    while True:
        ret, frame = vid.read()
        if not ret:
            print("Failed to capture video frame.")
            break

        # Process only every nth frame to reduce load
        global frame_count
        frame_count += 1
        if frame_count % process_every_n_frame != 0:
            continue

        # Resize frame to speed up recognition
        resized_frame = cv2.resize(frame, (640, 480))  # Reduce resolution for faster processing

        # Perform face recognition on the captured frame
        try:
            start_time = time.time()
            people = DeepFace.find(img_path=resized_frame, 
                                   db_path="/Users/komangandikawirasantosa/Face-Rec/DeepFace-IPYNB/Database", 
                                   model_name=models[0], 
                                   distance_metric=metrics[1], 
                                   enforce_detection=False)
            end_time = time.time()
            print(f"Recognition took {end_time - start_time:.2f} seconds")
        except Exception as e:
            print(f"Error during face recognition: {e}")
            continue

        # Check if the people list is empty
        if not people:
            cv2.putText(frame, "Unknown", (50, 50), cv2.FONT_ITALIC, 2, (0, 0, 255), 2)
        else:
            for person in people:
                if all(key in person for key in ['source_x', 'source_y', 'source_w', 'source_h']):
                    # Retrieve the coordinates of the face bounding box safely
                    x = person['source_x'][0] if len(person['source_x']) > 0 else 0
                    y = person['source_y'][0] if len(person['source_y']) > 0 else 0
                    w = person['source_w'][0] if len(person['source_w']) > 0 else 0
                    h = person['source_h'][0] if len(person['source_h']) > 0 else 0

                    # Draw a rectangle around the face
                    cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

                    # Get the person's name and display it on the image
                    if 'identity' in person and len(person['identity']) > 0:
                        identity = person['identity'][0].split('/')[-1].split('.')[0]  # Extract name without extension
                        cv2.putText(frame, identity, (x, y), cv2.FONT_ITALIC, 1, (0, 0, 255), 2)
                    else:
                        cv2.putText(frame, "Unknown", (x, y), cv2.FONT_ITALIC, 1, (0, 0, 255), 2)

        # Display the resulting frame
        cv2.namedWindow('frame', cv2.WINDOW_NORMAL)
        cv2.resizeWindow('frame', 960, 720)
        cv2.imshow('frame', frame)

        # Break loop on 'q' press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the video capture object and close windows
    vid.release()
    cv2.destroyAllWindows()

# Start real-time face recognition
realtime_face_recognition()
```
